```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/Salary_Data.csv')  # path depends on where you uploaded it
```

```python
data.head()      # Shows top 5 rows
data.shape       # Tells how many rows/columns
data.isnull().sum()  # Checks for missing values
```

```python
x = data.iloc[:, :1].values  # Years of Experience
y = data.iloc[:, 1:2].values  # Salary
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print("Predicted:\n", y_pred)
print("Actual:\n", y_test)
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('SALARY VS EXPERIENCE (training set)')
plt.xlabel('Experience in Years')
plt.ylabel('Salary in Rupees')
plt.show()
```

```python
plt.scatter(x_test, y_test, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')  # line from training model
plt.title('SALARY VS EXPERIENCE (testing set)')
plt.xlabel('Experience in Years')
plt.ylabel('Salary in Rupees')
plt.show()
```
