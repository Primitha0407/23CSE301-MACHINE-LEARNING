```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/content/Taxi_Trips_Sample.csv')
data.head()
```

```python
x = data.drop('Fare', axis=1).values  # All features except fare
y = data['Fare'].values.reshape(-1, 1)  # Target = Fare
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

# Compare actual vs predicted
comparison = pd.DataFrame({'Actual Fare': y_test.flatten(), 'Predicted Fare': y_pred.flatten()})
print(comparison)
```

```python
plt.scatter(y_test, y_pred, color='blue')
plt.xlabel("Actual Fare")
plt.ylabel("Predicted Fare")
plt.title("Actual vs Predicted Taxi Fare")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')  # Line of perfect prediction
plt.show()
```
